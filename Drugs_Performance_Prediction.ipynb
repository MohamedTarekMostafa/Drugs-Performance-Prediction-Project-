```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Drugs_Performance.csv")
df.head()
```

```python

x = df[["Age"]].values
y = df[["Drugs_Perf"]].values
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel("Age")
plt.ylabel("Drugs_Perf")
plt.show()
##can,t fit this data with linear regression model
###Fitting with polynomial model
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
poly = PolynomialFeatures(degree = 2)
x_poly = poly.fit_transform(x_train)
poly.fit(x_poly, y_train)
```

```python

model = LinearRegression()
model.fit(x_poly,y_train)
```

```python

plt.scatter(x,y, color = 'orange')
plt.plot(x,model.predict(poly.fit_transform(x)), color = 'black')
plt.title('Polynomial Regression')
plt.xlabel('Age')
plt.ylabel('Drugs Performance')
```

```python
print('Prediction by the model:\t', model.predict(poly.fit_transform(x_test)))
```
